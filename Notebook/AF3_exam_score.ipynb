# AF3 – Predicción de calificaciones con Regresión Lineal
Notebook listo para ejecutar paso a paso.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from pathlib import Path
import numpy as np
```

## 1. Cargar dataset

```python
ruta = Path('data/exam_score_prediction.csv')
df = pd.read_csv(ruta)
df.head()
```

## 2. Preprocesamiento

```python
df = df.dropna(how='all')
for col in df.select_dtypes(include=['int64','float64']).columns:
    if df[col].isnull().any():
        df[col].fillna(df[col].mean(), inplace=True)

y = df.iloc[:, -1]
X = df.iloc[:, :-1]
X = pd.get_dummies(X, drop_first=True)
X.head()
```

## 3. División y normalización

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled[:5]
```

## 4. Entrenamiento del modelo

```python
modelo = LinearRegression()
modelo.fit(X_train_scaled, y_train)
y_pred = modelo.predict(X_test_scaled)
y_pred[:5]
```

## 5. Evaluación de métricas

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mse, mae, r2
```

## 6. Gráfica valores reales vs predichos

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.xlabel('Real')
plt.ylabel('Predicho')
plt.title('Regresión Lineal – Valores Reales vs Predichos')
min_val = min(min(y_test), min(y_pred))
max_val = max(max(y_test), max(y_pred))
plt.plot([min_val, max_val], [min_val, max_val], 'r')
plt.tight_layout()
plt.savefig('imagenes/exam_scatter.png', dpi=300)
plt.show()
```
